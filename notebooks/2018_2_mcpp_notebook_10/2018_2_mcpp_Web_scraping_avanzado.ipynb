-  **OJO! en este tipo de páginas web, utilizar solamente *Beautiful Soup* no es suficiente a la hora de hacer Web scraping**


-  Cuando una web está escrita en JavaScript el buscador realiza un proceso llamado ***Rendering*** que consiste en *ejecutar* el código JavaScript de la página


-  ¿Cuál es el problema?... SOLO los navegadores pueden hacer ***Rendering*** (Chrome, Firefox, Safari, Explorer)
 
 
-  ***Beautiful Soup*** y ***urllib*** NO le permiten a Python realizar este proceso


-  Es decir, el código HTML retornado estará incompleto y probablemente la información que necesitemos no estará presente

## ¿Cómo hacer web scraping de páginas escritas en JavaScript?



### Lo que necesitaremos:

- Python 3


- Beautiful Soup


- [Selenium](http://selenium-python.readthedocs.io/)


- [PhantomJS](http://phantomjs.org/)

```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.firefox.firefox_binary import FirefoxBinary

# Definimos nuestro url


url = 'https://www.mercadoni.com.co/#!/'
```

Veamos que pasaría si utilizamos urllib y Beautiful Soup

```python
from urllib.request import urlopen


webpage = urlopen(url) 

soup = BeautifulSoup(webpage, "html.parser")

print(soup.prettify)
```

Nada parecido a lo que queremos encontrar...

Intentemos otro link de la misma página:

```python
url = 'https://www.mercadoni.com.co/#!/tienda/jumbo-colombia/categoria/novedades-alimentos'



webpage = urlopen(url) 

soup = BeautifulSoup(webpage, "html.parser")

print(soup.prettify)
```

```python
import time

url = 'https://www.mercadoni.com.co/#!/tienda/olimpica/categoria/frutas-y-verduras'

#browser = webdriver.PhantomJS(executable_path=r"C:\bin\phantomjs\bin\phantomjs.exe")
browser = webdriver.Chrome(executable_path=r"C:\\Users\jkp28\Downloads\chromedriver")
  
browser.get(url)
    
time.sleep(10)
```

Acabamos de abrir el url con el buscador **Phantom** por medio de Python

```python
html = browser.execute_script("return document.getElementsByTagName('html')[0].innerHTML")
    
soup = BeautifulSoup(html, "html.parser")

```

**Cuando usamos la función execute_script, Python hace el rendering de la página por medio de Phantom**

```python
print(soup.prettify)
```

Ahora sí estamos capturando el código HTML que nos interesa

```python
#Listemos los productos para ver su información

items = soup.findAll(name = 'li', class_ = 'product-item ng-scope')

for item in items:

    marca = item.meta.get('content')

    try:
        precio = item.find(name = 'span', class_ = 'price ng-binding ng-scope').string
        precio_descuento = 'NA'
    except:
        precio = item.find(name = 'span', class_ = 'special-price ng-binding ng-scope').string
        precio_descuento = item.find(name = 'span', class_ = 'price with-special-price ng-binding ng-scope').string

    nombre = item.find(name = 'div', class_ = 'name ng-binding').string
    unidades = item.find(name = 'span', class_ = 'unit ng-binding').string
    
    print(nombre)
    print(marca)
    print(unidades)
    print(precio)
    print(precio_descuento)
    print('------------------')
    print('------------------')
    print('------------------')
```
