### Un ejemplo a partir de un TXT

```python
with open('acuerdo_farc.txt') as f:
    acuerdo = f.read()
```

```python
print(acuerdo)
```

```python
type(acuerdo)
```

```python
all_tokens = acuerdo.split()
all_tokens
```

```python
# Dejar caracteres alfab√©ticos
alpha = [w for w in all_tokens if w.isalpha()]
alpha
```

¬øY las palabras con tilde? El anterior m√©todo es muy √∫til en ingl√©s, no en espa√±ol :(

```python
import string
# Grab punctuation characters

string.punctuation
```

```python
# Add additional relevant found characters to strip punctuation

punctuation = string.punctuation + '‚Äì¬°¬ø‚Äù‚Äú‚Ä¢\r¬¥'
punctuation
```

```python
# Strip punctuation from string

def no_punct(string):
    transtable = string.maketrans('', '', punctuation)
    return string.translate(transtable)
```

```python
tokens_no_punc = list(map(no_punct, all_tokens))
tokens_no_punc
```

```python
# Todo en min√∫sculas
lower = [w.lower() for w in tokens_no_punc]
lower
```

```python
# Quitar n√∫meros
alpha = [w for w in lower if not w.isdigit()]
alpha
```

```python
len(alpha)
```

```python
len(set(alpha))
```

```python
freqdist = nltk.FreqDist(alpha)
freqdist
```

```python
freqdist.most_common()
```

```python
from nltk.corpus import stopwords
# Grab stopwords in Spanish
stopwords_esp = stopwords.words('spanish')
stopwords_esp
```

```python
# Filter list of all tokens based on list of words of interest

tokens_no_stop = [w for w in alpha if w not in stopwords_esp]
tokens_no_stop
```

```python
freqdist = nltk.FreqDist(tokens_no_stop)
freqdist.most_common()
```

```python
# From inspection of frequency distribution, make list of words of interest

not_of_interest = ['','p√°gina','mmv','as√≠','zvtn','da']
```

```python
# Filter list of all tokens based on list of words of interest

tokens_of_interest = [w for w in tokens_no_stop if w not in not_of_interest]
tokens_of_interest
```

```python
# Si queremos aplicar m√©todos de objetos nltk.Text:
tokens_nltk = nltk.Text(tokens_of_interest)
tokens_nltk
```

Nube de palabras:

```python
# Join list in string

tokens_of_interest_str = ' '.join(tokens_of_interest)
tokens_of_interest_str
```

```python
# Write to txt file

with open('tokens_of_interest_str.txt', 'a') as f:
    f.write(tokens_of_interest_str)
```

```python
# Based on https://github.com/amueller/word_cloud/blob/master/examples/masked.py
# Word cloud requirements

from os import path
from wordcloud import WordCloud

d = path.dirname('/Users/Santiago/Dropbox/Clases/URosario/2018-II/repo_mcpp_2018_2/notebooks/tokens_of_interest_str.txt')

# Read the whole text
text = open(path.join(d, 'tokens_of_interest_str.txt')).read()

wordcloud = WordCloud().generate(text)

# Display the generated image:
# the matplotlib way:
import matplotlib.pyplot as plt
plt.imshow(wordcloud)
plt.axis("off")

# take relative word frequencies into account, lower max_font_size
wordcloud = WordCloud(max_font_size=40, relative_scaling=.5).generate(text)
plt.figure()
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
