```python
m1 = [[0,1,4], [9,16,25], [36,49,64], [81,100,121]]
```

```python
print(m1)
```

```python
print(m1*2)
```

```python
m2 = []
for i in range(len(m1)):
    l = []
    for j in range(len(m1[0])):
        l.append(m1[i][j]*2)
    m2.append(l)
print(m2)
```

```python
m2 = [[x*2 for x in row] for row in m1]
print(m2)
```

```python
plt.title("My graph")
plt.xlabel("Some x values")
plt.ylabel("Some y values")
plt.axhline(0, color="gray", linestyle="--")
plt.axvline(0, color="gray", linestyle="--")
plt.xlim(-60, 60)
plt.ylim(-30, 30)
plt.plot(xs,ys, label = "My values")
plt.legend();
```

```python
with plt.xkcd():
    plt.title("My graph")
    plt.xlabel("Some x values")
    plt.ylabel("Some y values")
    plt.axhline(0, color="gray", linestyle="--")
    plt.axvline(0, color="gray", linestyle="--")
    plt.xlim(-60, 60)
    plt.ylim(-30, 30)
    plt.plot(xs,ys, label = "My values")
    plt.legend()
```

```python
xs = np.linspace(-20, 20, 200)
```

```python
labels = ["Corn", "Soybeans", "Wheat", "Alfalfa", "Cotton"]
values = [24.4, 17.7, 8.6, 8.3, 6.1]
          
plt.title("Major Crops in the USA – 1997")
plt.ylabel("Value of Production (in US$ billions)")

xs = np.arange(len(labels))

# Add labels (or "ticks") to X axis
xtick_locs = xs + 0.4
plt.xticks(xtick_locs, labels)

plt.bar(xs, values);
```
